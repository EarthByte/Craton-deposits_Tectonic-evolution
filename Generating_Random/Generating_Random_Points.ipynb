```python
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely.geometry import Point
import random

# Function to generate random points uniformly within a polygon
def generate_random_points_uniform(polygon, num_points):
    points = []
    minx, miny, maxx, maxy = polygon.bounds
    while len(points) < num_points:
        # Generate random x and y within the bounding box of the polygon
        random_point = Point(random.uniform(minx, maxx), random.uniform(miny, maxy))
        # Check if the random point lies within the polygon
        if polygon.contains(random_point):
            points.append(random_point)
    return points

# Load your polygon shapefile
polygon_gdf = gpd.read_file('./Datasets/Buffer186km.shp')

# Extract the polygon geometry (assuming it's the first geometry in the shapefile)
polygon = polygon_gdf.geometry[0]

# Generate 500 random points uniformly inside the polygon
num_points = 500
random_points = generate_random_points_uniform(polygon, num_points)

# Create a GeoDataFrame from the random points
gdf = gpd.GeoDataFrame(geometry=random_points, crs=polygon_gdf.crs)

# Add 180 columns named age_0010 to age_1800, filled with values from 10 to 1800
for i in range(1, 181):
    column_name = f'age_{i:03d}0'  # Creates column names like age_0010, age_0020, ..., age_1800
    gdf[column_name] = i * 10

# Save the result as a new shapefile
output_path = './Datasets/random_points_with_ages.shp'
gdf.to_file(output_path)

print(f"Random points with age columns saved to: {output_path}")
```
